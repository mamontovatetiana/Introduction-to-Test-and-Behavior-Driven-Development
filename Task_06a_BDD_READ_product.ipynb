# Task 6a: BDD Scenario for Reading a Product

**Caption:** This BDD scenario verifies retrieving a product by ID and checking its fields.

```
Scenario: Read a Product
    Given the following products
        | name | description | price | available | category |
        | Hat  | A red hat   | 10.00 | true      | CLOTHS   |
    When I visit the "Home Page"
    And I set the "Name" to "Hat"
    And I press the "Search" button
    Then I should see the message "Success"
    And I copy the "Id" field
    When I press the "Clear" button
    And I paste the "Id" field
    And I press the "Retrieve" button
    Then I should see the message "Success"
    And I should see "Hat" in the "Name" field
    And I should see "A red hat" in the "Description" field
    And I should see "10.00" in the "Price" field
    And I should see "true" in the "Available" field
    And I should see "CLOTHS" in the "Category" field
```
