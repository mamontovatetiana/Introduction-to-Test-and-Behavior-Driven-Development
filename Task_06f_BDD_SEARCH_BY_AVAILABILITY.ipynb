# Task 6f: BDD Scenario for Searching by Availability

**Caption:** This BDD scenario verifies filtering products by availability.

```
Scenario: Search Products by Availability
    Given the following products
        | name  | description | price | available | category    |
        | Hat   | A red hat   | 10.00 | true      | CLOTHS      |
        | Phone | Old phone   | 50.00 | false     | ELECTRONICS |
    When I visit the "Home Page"
    And I press the "Clear" button
    Then I should see the message "Success"
    When I select "true" in the "Available" dropdown
    And I press the "Search" button
    Then I should see the message "Success"
    And I should see "Hat" in the results
    And I should see "true" in the results
    And I should not see "Phone" in the results
```
