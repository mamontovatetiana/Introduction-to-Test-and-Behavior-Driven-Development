# Task 7c: Text Not Present Step Definition

**Caption:** This Selenium step definition verifies that a specific text does not appear in search results.

```
@then('I should not see "{text}" in the results')
def step_impl(context, text):
    WebDriverWait(context.driver, context.wait_seconds).until_not(
        EC.text_to_be_present_in_element((By.ID, "search_results"), text)
    )
```
