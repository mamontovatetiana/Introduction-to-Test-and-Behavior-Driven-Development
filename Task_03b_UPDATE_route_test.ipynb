# Task 3b: UPDATE Product Route Test

**Caption:** This route test verifies that the API can update an existing product.

```
def test_update_product(self):
    """It should Update an existing Product"""
    product = ProductFactory()
    response = self.client.post(BASE_URL, json=product.serialize(), content_type="application/json")
    self.assertEqual(response.status_code, status.HTTP_201_CREATED)

    new_product = response.get_json()
    new_product["description"] = "unknown"

    response = self.client.put(
        f"{BASE_URL}/{new_product['id']}",
        json=new_product,
        content_type="application/json"
    )

    self.assertEqual(response.status_code, status.HTTP_200_OK)

    updated_product = response.get_json()
    self.assertEqual(updated_product["description"], "unknown")
```
