# Task 3f: LIST BY CATEGORY Product Route Test

**Caption:** This route test verifies that the API can list products filtered by category.

```
def test_query_product_by_category(self):
    """It should Query Products by Category"""
    self._create_products(10)

    response = self.client.get(
        BASE_URL,
        query_string="category=CLOTHS"
    )

    self.assertEqual(response.status_code, status.HTTP_200_OK)

    data = response.get_json()
    self.assertGreater(len(data), 0)
    for product in data:
        self.assertEqual(product["category"], "CLOTHS")
```
