# Task 3g: LIST BY AVAILABILITY Product Route Test

**Caption:** This route test verifies that the API can list products filtered by availability.

```
def test_query_product_by_availability(self):
    """It should Query Products by Availability"""
    self._create_products(10)

    response = self.client.get(
        BASE_URL,
        query_string="available=true"
    )

    self.assertEqual(response.status_code, status.HTTP_200_OK)

    data = response.get_json()
    self.assertGreater(len(data), 0)
    for product in data:
        self.assertEqual(product["available"], True)
```
